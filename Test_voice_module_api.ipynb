## Khám phá Gemini API
Gemini API cấp cho bạn quyền truy cập vào các mô hình Gemini do Google DeepMind tạo. Các mô hình Gemini được xây dựng từ đầu theo hướng đa phương thức, vì vậy, bạn có thể suy luận liền mạch trên văn bản, hình ảnh, mã và âm thanh.

**Cách bắt đầu**
*  Truy cập vào <a href="https://aistudio.google.com/">Google AI Studio</a> rồi đăng nhập bằng Tài khoản Google của bạn.
*  <a href="https://aistudio.google.com/app/apikey">Tạo một khoá API</a>.
* Sử dụng hướng dẫn bắt đầu nhanh dành cho <a href="https://colab.research.google.com/github/google-gemini/cookbook/blob/main/quickstarts/Get_started.ipynb">Python</a> hoặc dùng <a href="https://colab.research.google.com/github/google-gemini/cookbook/blob/main/quickstarts/rest/Prompting_REST.ipynb">curl</a> để gọi API REST.

**Khám phá các tính năng nâng cao của Gemini**
*  Thử nghiệm với <a href="https://colab.research.google.com/github/google-gemini/cookbook/blob/main/quickstarts/Image-out.ipynb">kết quả đầu ra đa phương thức</a> của Gemini, kết hợp văn bản và hình ảnh theo cách lặp lại.
*  Khám phá <a href="https://colab.research.google.com/github/google-gemini/cookbook/blob/main/quickstarts/Get_started_LiveAPI.ipynb">API Trực tiếp đa phương thức</a> &#40;bản minh hoạ có <a href="https://aistudio.google.com/live">tại đây</a>&#41;.

**Tìm hiểu các trường hợp sử dụng phức tạp**

*  Dùng cửa sổ ngữ cảnh lớn của Gemini và Imagen để tạo <a href="https://colab.research.google.com/github/google-gemini/cookbook/blob/main/examples/Book_illustration.ipynb">hình minh hoạ dựa trên toàn bộ cuốn sách</a>.
